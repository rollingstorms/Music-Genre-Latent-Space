```python
from pyarrow import feather
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tracks_df = feather.read_feather('data/all_tracks.feather')
```

```python
tracks_df = tracks_df.drop_duplicates(subset=['track_id']).reset_index()
```

```python
tracks_df['release_date'] = pd.to_datetime(tracks_df['release_date'],  errors='coerce', infer_datetime_format=True)
```

```python
tracks_df['year'] = tracks_df['release_date'].dt.year
```

```python
tracks_df.isna().sum()
```

```python
tracks_df.describe()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
tracks_df.hist(ax=ax, bins=50)
plt.show()
```
