```python
from pyarrow import feather
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tracks_df = feather.read_feather('data/all_tracks.feather')
```

```python
tracks_df = tracks_df.drop_duplicates(subset=['track_id']).reset_index(drop=True)
```

```python
tracks_df['release_date'] = pd.to_datetime(tracks_df['release_date'],  errors='coerce', infer_datetime_format=True)
```

```python
tracks_df['year'] = tracks_df['release_date'].dt.year
```

```python
fig, ax = plt.subplots(figsize=(15,5))
tracks_df.hist(ax=ax, bins=50)
plt.show()
```

```python
tracks_df.sort_values(by='track_popularity', ascending=False)
```

```python
all_genres = set()
for genres in tracks_df['artist_genres']:
    for genre in genres:
        all_genres.add(genre)
```

```python
all_genres_dict = dict.fromkeys(all_genres, 0)
for genres in tracks_df['artist_genres']:
    for genre in genres:
        all_genres_dict[genre] += 1
```

```python
all_genres_dict_sort = {k: v for k, v in sorted(all_genres_dict.items(), key=lambda item: item[1] , reverse=True) if v > 50}
```

```python
len(all_genres_dict_sort)
```

```python
results = []
for idx, row in tracks_df.iterrows():
    result = {'id':row.artist_id}
    for genre in row.artist_genres:
        result[genre] = 1
    results.append(result)
    
```
