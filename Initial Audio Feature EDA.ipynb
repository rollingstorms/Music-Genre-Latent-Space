```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
signal, sr = librosa.load('data/preview_test.mp3')
```

```python
plt.figure(figsize=(15,5))
plt.imshow(librosa.power_to_db(librosa.feature.melspectrogram(y=signal, sr=sr), ref=np.max), aspect=2)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(librosa.power_to_db(librosa.feature.mfcc(y=signal, sr=sr), ref=np.max), aspect=10)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(librosa.feature.chroma_cqt(y=signal, sr=sr), aspect=20)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(librosa.feature.zero_crossing_rate(y=signal), aspect=50)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(librosa.feature.spectral_rolloff(y=signal), aspect=50)
```
