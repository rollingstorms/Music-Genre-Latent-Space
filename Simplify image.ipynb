```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import numpy.ma as ma
```

```python
img = tf.keras.preprocessing.image.load_img('data/Spotify/comp_pngs/0t03JYy5TL5UrIKuwqA5XF.png', color_mode='rgb')
img
```

```python
scale = 1./255
img = scale*np.array(img)
plt.imshow(img)
```

```python
plt.imshow(img[:,:,0])
img = img[:,:,0]
```

```python
img.shape
```

```python
img_min = img.min()
img_max = img.max()
simplify_threshold = .5
threshold = simplify_threshold * (img_max - img_min) + img_min
mx = ma.masked_where(img < threshold, img, copy=True)
img = (mx.filled(fill_value=threshold) - threshold)
img = img  * (img_max / img.max())
```
